<h2 id="correlation_causation">5. Correlation and Causation</h2>

```python
from scipy import stats
```

<h3>Let's calculate the Pearson Correlation Coefficient and P-value of 'wheel-base' and 'price'.</h3>

```python
pearson_k,p_value=stats.pearsonr(df['wheel-base'], df['price'])
pearson_k,p_value
```

<h5> coefficient is 0.5  so moderate correlation and p value is lessthan 0.001 so it has moderate correlation, so the linear relation ship is not signifigently strong</h5>

<h3>Let's calculate the Pearson Correlation Coefficient and P-value of 'horsepower' and 'price'.</h3>

```python
pearson_k1,p_value1=stats.pearsonr(df['horsepower'], df['price'])
print(" Pearson Cofficient:" ,pearson_k1 ," with P value : " ,p_value1)
```

```python
pearson_coef, p_value = stats.pearsonr(df['length'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value)  
```

<h2 id="anova">6. ANOVA Analysis of Variance</h2>

```python
df_test1 =df[['drive-wheels','price']]
df_grp1 =df_test1.groupby(['drive-wheels'])
df_grp1.head(2)                        
```

```python
df_grp1.get_group('4wd')['price']
```

```python
f_val, p_val = stats.f_oneway(df_grp1.get_group('fwd')['price'], df_grp1.get_group('rwd')['price'], df_grp1.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```

<h3>Separately: fwd and rwd</h3>

```python
f_val, p_val = stats.f_oneway(df_grp1.get_group('fwd')['price'], df_grp1.get_group('rwd')['price'])
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```

<h3> seperately:4wd and rwd </h3>

```python
f_val, p_val = stats.f_oneway(df_grp1.get_group('4wd')['price'], df_grp1.get_group('rwd')['price'])
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```

<h4>4wd and fwd</h4>

```python
f_val, p_val = stats.f_oneway(df_grp1.get_group('4wd')['price'], df_grp1.get_group('fwd')['price'])
print( "ANOVA results: F=", f_val, ", P =", p_val) 
```
